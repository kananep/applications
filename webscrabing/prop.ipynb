```python
bedroom_numbers = []

for i, item in enumerate(all[3].find('span', {'class': 'd-inline-flex ml-2'})):
    bedroom_text = item.text.strip()
    if i == 0 and '' in bedroom_text:
        bedroom_numbers.append(None)
    else:
        bedroom_numbers.append(bedroom_text)

# Now, bedroom_numbers contains the bedroom numbers including None for the first iteration
print(bedroom_numbers)
```

```python

lst = []
for info in all:
    
    dictionary = {}
    
    input_string = info.find('a',{'class':'text-dark'}).text.replace('\n',' ').replace(' ','')
    dictionary['Title'] = re.sub(r'(?<=\w)([A-Z])', r' \1', input_string)
    dictionary['Price'] = info.find('div',{'class':'col-md-auto col'}).text.replace('\n',' ').replace(' ','')
    dictionary['Locality']  = [span.text for span in info.find_all('span', class_=False)][1]
    dictionary['Posted on'] = info.find('span',{'class':'text-dark'}).text

    
    dictionary['Bedroom'] = info.find('span',{'class':'d-inline-flex ml-2'}).text  
    
    
    try:
        dictionary['Bathroom'] = info.find(class_="icon-bath").find_next_sibling('span').text.strip()
    except: 
        pass  
    
    try:
        dictionary['Square Feet'] = ''.join(filter(str.isdigit, info.find(class_="icon-sqft").find_next_sibling('span').text.strip()))
    except:
        pass    
    
    lst.append(dictionary)
```

```python
modified_data = []
for i, item in enumerate(lst):
    # Check if it's the first iteration and if '850  ' is present in the 'Bedroom' value
    if i == 0 and '850  ' in item['Bedroom']:
        # If so, append a modified dictionary with 'Bedroom' set to None
        modified_data.append({key: (None if key == 'Bedroom' else value) for key, value in item.items()})
    else:
        # Otherwise, append the original dictionary
        modified_data.append(item)

```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

base_url = 'https://www.prophuntgh.com/properties-for-sale-in-accra/srp/page/'

l = []  # Initialize an empty list outside the loop

for page in range(8, 10):  # Iterating over pages 8 and 9
    print(base_url + str(page))
    r = requests.get(base_url + str(page))
    content = r.content
    soup = BeautifulSoup(content, 'html.parser')
    all = soup.find_all('div', {'class': 'col-12 mb-md-4 mb-3 pb-lg-1'})

    for info in all:
        dictionary = {}

        input_string = info.find('a', {'class': 'text-dark'}).text.replace('\n', ' ').replace(' ', '')
        dictionary['Title'] = re.sub(r'(?<=\w)([A-Z])', r' \1', input_string)
        dictionary['Price'] = info.find('div', {'class': 'col-md-auto col'}).text.replace('\n', ' ').replace(' ', '')
        dictionary['Locality'] = [span.text for span in info.find_all('span', class_=False)][1]
        dictionary['Posted on'] = info.find('span', {'class': 'text-dark'}).text
        try:
            dictionary['Bedroom'] = info.find('span', {'class': 'd-inline-flex ml-2'}).text
        except:
            pass
        try:
            dictionary['Bathroom'] = info.find(class_="icon-bath").find_next_sibling('span').text.strip()
        except:
            pass
        try:
            dictionary['Land Square Feet'] = info.find(class_='icon-sqft').find_next_sibling('span').text.strip()
        except:
            pass

        l.append(dictionary)  # Append the dictionary inside the loop

# Create DataFrame after the loop to include data from all pages
df = pd.DataFrame(l)
```
