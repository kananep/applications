```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://www.century21.com/real-estate/los-angeles-ca/LCCALOSANGELES/?pdp=REN0'

for page in range(28403880,28403900,1):
    print(base_url+str(page))
    r = requests.get(base_url+str(page))
    content = r.content
    soup = BeautifulSoup(content,'html.parser')
    #print(soup.prettify())
    all = soup.find_all('div',{'data-brand-cd':'REN'})
    print(all)
```
